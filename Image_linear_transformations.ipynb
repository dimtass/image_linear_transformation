Use linear transformation on images
----

This is an example on how to use linear transformation on bpm images.
For this you need to have a bmp file named `test.bpm` which is 320x240 px


```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
```

## Read the bmp file

Read the bmp file and then plot it to view the image

```python
img = imageio.imread('test.bmp')
```

```python
plt.imshow(img)
```

## Create the transformation vector

We now create a transformation vector:

```python
Tr = np.array([[0, 1, 0], [-1, 0, 0], [0, 0, 1]])
Tr
```

## Transformation function

```python
def Transform(img_array, img_transformed, transformation_array):
    for i, row in enumerate(img_array):  
        for j, col in enumerate(row):
            pixel_data = img_array[i, j, :]
            input_coords = np.array([i, j, 1])
            i_out, j_out, _ = transformation_array @ input_coords
            img_transformed[i_out, j_out, :] = pixel_data
```

## Apply transformation

```python
new_img = np.empty((320, 240, 3), dtype=np.uint8)
Transform(img, new_img, Tr)

plt.figure(figsize=(5, 5))  
plt.imshow(new_img)
```

## Save the image

```python
from PIL import Image
new_bpm = Image.fromarray(new_img)
new_bpm.save('result.bmp')
```

## Notes

You can trim the resulted image size as you expect what size you going to get.
